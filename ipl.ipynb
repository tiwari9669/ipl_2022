1. which team won the most matches?


```python
match_wins = df['match_winner'].value_counts()
sns.barplot(x=match_wins.values, y = match_wins.index, palette='rainbow')
plt.title("Number of matches won by each team")
plt.xlabel("Number of matches")
plt.ylabel("Team")
```

2. Toss Decision Trend

```python
sns.countplot(x=df['toss_decision'], palette='rainbow') # hrer show the around 80% who won the toss they decide to feild first
plt.title("Toss Decision Trend")
```

```python
import plotly.express as px ## use of plotly on the same question to more attractive
import pandas as pd

toss_counts = df.groupby(by=['toss_decision']).size().reset_index(name='count')

fig = px.bar(data_frame=toss_counts, x='toss_decision', y='count', title='Toss Decision Trend')

fig.show()
```

3. Toss winner vs Match winner

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
perc = (count*100)/df.shape[0]
perc
```

4. How do teams win ?(Runs or Wicket)

```python
sns.countplot(x=df['won_by'])
plt.title("How do teams win ?")
```

2 Top scorers

```python
high = df.groupby(by='top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high.plot(kind = 'barh')
```

10 Best boling figure

```python
df.head()
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind = 'barh')
```
